## Histogram
- Histograms are used to analyse one variable/feature at a time. We use histo when we want to see the count of the variable on a chart

```python
plt.hist(df[["No. of Clicks"]])
plt.title("Histogram of number of clicks")
plt.xlabel("no of clicks")
plt.ylabel("count of clicks")
plt.show()
```

```python
sns.distplot(df["No. of Clicks"],bins=10)
```

```python
sns.distplot(df["No. of Clicks"],bins=14)
```

```python
sns.histplot(df["No. of Clicks"])
plt.title("Histogram of number of clicks")
plt.xlabel("no of clicks")
plt.ylabel("count of clicks")
plt.show()
```

```python
df["No. of Clicks"].plot(kind="hist")
```

```python
plt.hist(df[["Pages Viewed"]])
plt.title("Histogram Viewed Pages")
plt.xlabel("Number of Pages Viewed")
plt.ylabel("count of Pages Viewed")
plt.show()
```

```python
sns.distplot(df["Pages Viewed"],bins=15)
```

```python
df["Pages Viewed"].plot(kind="hist")
```

```python
plt.hist(df[["Time (min)"]])
plt.title("Histogram of time spent by the users in minute")
plt.xlabel("time in minutes")
plt.ylabel("count of time spent by the user")
plt.show()
```

## Pie Chart
- Pie chart is a circular chart which is divided into slices 
- it is used to view and compare categorical column

```python
df["Browser"].unique()
```

```python
df["Browser"].value_counts()
```

```python
plt.pie(df["Browser"].value_counts(),labels=['Chrome', 'Internet Explorer', 'Firefox'],autopct="%.2f%%")
plt.show()
```

```python
colors=sns.color_palette("bright")
plt.pie(df["Browser"].value_counts(),labels=['Chrome', 'Internet Explorer', 'Firefox'],colors=colors,autopct="%.2f%%")
plt.show()
```

```python
df["Browser"].value_counts().plot(kind="pie",autopct="%.2f",colors=colors)
```

## Bar Plot / Bar Graph
- A bar plot is used to analyse several group of data in a column

```python
Day_value_counts=df["Day"].value_counts()
```

```python
Day_value_counts
```

```python
Day_value_counts.index
```

```python
Day_value_counts.values
```

```python
plt.figure(figsize=(20,5))   # figsize -- (width,height)
plt.bar(Day_value_counts.index,Day_value_counts.values)
plt.title("A")
```

```python
df["Day"].value_counts().plot(kind="bar")
```
